```python
import random
import time

# 0 = sạch, 1 = bẩn
room = [
    [1, 0, 1],
    [1, 1, 0],
    [0, 1, 1]
]

# vị trí robot
robot_pos = [1, 1]

def print_room(room, robot_pos):
    for i in range(3):
        for j in range(3):
            if [i, j] == robot_pos:
                print("R", end=" ")
            else:
                print(room[i][j], end=" ")
        print()
    print("------")

def get_actions(row, col):
    actions = []
    if row > 0:
        actions.append("UP")
    if row < 2:
        actions.append("DOWN")
    if col > 0:
        actions.append("LEFT")
    if col < 2:
        actions.append("RIGHT")
    return actions

def move(robot_pos, action):
    row, col = robot_pos
    if action == "UP":
        return [row-1, col]
    elif action == "DOWN":
        return [row+1, col]
    elif action == "LEFT":
        return [row, col-1]
    elif action == "RIGHT":
        return [row, col+1]

# chạy random
for step in range(15):
    row, col = robot_pos

    print(f"Step {step+1}")
    print_room(room, robot_pos)

    # nếu có bụi → hút
    if room[row][col] == 1:
        print("Action: SUCK")
        room[row][col] = 0
    else:
        actions = get_actions(row, col)
        action = random.choice(actions)
        print("Action:", action)
        robot_pos = move(robot_pos, action)

    time.sleep(0.5)
```
