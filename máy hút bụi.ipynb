```python
import random
matran = [
    ['.', '.', '.', 'T'],
    ['X', 'X', '.', '.'],
    ['.', '.', 'T', '.'],
    ['.', '.', '.', '.']
]
x, y = 0, 0
moves = [
    (-1, 0, "len"),
    (1, 0, "xuong"),
    (0, -1, "trai"),
    (0, 1, "phai")
]

def display():
    temp = [row[:] for row in matran]
    temp[x][y] = 'R'
    for row in temp:
        print(' '.join(row))
    print()

def valid(nx, ny):
    if nx < 0 or ny < 0 or nx >= 4 or ny >= 4:
        return False
    if matran[nx][ny] == 'X':
        return False
    return True

print("R là máy hút bụi")
print("X là vật cản")
print("T là rác")
print(". là ô đã sạch")
print("Bản đồ ban đầu:")
display()
solan = 0

while 'T' in str(matran):
    if matran[x][y] == 'T':
        print("Đang hút rac tại:", x, y)
        matran[x][y] = '.'
    random.shuffle(moves)
    for dx, dy, huong in moves:
        nx = x + dx
        ny = y + dy
        if valid(nx, ny):
            x = nx
            y = ny
            solan += 1
            print("Bước", solan, ": đi", huong)
            print("bản đồ sau khi đi chuyển:")
            display()
            break

print("Đã hút sạch rác sau", solan, "bước.")
```
