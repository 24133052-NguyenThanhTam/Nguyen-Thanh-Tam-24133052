```python
KETQUA = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
MA_TRAN = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]
def tim_o_trong(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j
def so_lan_di(board):
    dem = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] != 0 and board[i][j] != KETQUA[i][j]:
                dem += 1
    return dem
def in_ma_tran(matran):
    for row in matran:
        for x in row:
            if x == 0:
                print("0", end=" ")
            else:
                print(x, end=" ")
        print()
    print()
def kiem_tra(x, y):
    return 0 <= x < 3 and 0 <= y < 3
moves = [
    (-1, 0, "len"),
    (1, 0, "xuong"),
    (0, -1, "trai"),
    (0, 1, "phai")
]
SO_LAN = 0
print("Trang thai ban dau:")
in_ma_tran(MA_TRAN)
while MA_TRAN != KETQUA:
    x, y = tim_o_trong(MA_TRAN)
    best_h = 999
    best_move = None
    for dx, dy, action in moves:
        nx = x + dx
        ny = y + dy
        if kiem_tra(nx, ny):
            new_board = [row[:] for row in MA_TRAN]
            new_board[x][y], new_board[nx][ny] = \
                new_board[nx][ny], new_board[x][y]
            h = so_lan_di(new_board)
            if h < best_h:
                best_h = h
                best_move = (nx, ny, action)
    nx, ny, action = best_move
    MA_TRAN[x][y], MA_TRAN[nx][ny] = \
        MA_TRAN[nx][ny], MA_TRAN[x][y]
    SO_LAN += 1
    print("di chuyen lan", SO_LAN)
    print("hanh dong: di", action)
    print("ma tran sau khi di chuyen:")
    in_ma_tran(MA_TRAN)
print("da dat yeu cau bai toan trong", SO_LAN, "buoc di")
```
