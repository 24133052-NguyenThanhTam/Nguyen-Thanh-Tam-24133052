```python
import random
matran = [
    ['1', '0', '1', '1'],
    ['1', '0', '0', '0'],
    ['1', '0', '0', '0'],
    ['1', '0', '1', '1']
]
x, y = 0, 0
model = [[None for _ in range(4)] for _ in range(4)]
def display():
    vitri = [row[:] for row in matran]
    vitri[x][y] = 'R'
    for row in vitri:
        print(' '.join(row))
    print()
def get_actions(x, y):
    actions = []
    if x > 0:
        actions.append(("len", x - 1, y))
    if x < 3:
        actions.append(("xuong", x + 1, y))
    if y > 0:
        actions.append(("trai", x, y - 1))
    if y < 3:
        actions.append(("phai", x, y + 1))
    return actions
print("R : may hut bui")
print("1 : noi chua lam sach")
print("0 : khong con bui")
print()
print("ma tran ban dau:")
display()
solan = 0
while '1' in str(matran):
    percept = matran[x][y]
    print("vi tri hien tai:", (x, y))
    model[x][y] = percept
    print("Model hien tai:")
    for row in model:
        print(row)
    print()
    if percept == '1':
        print("may hut bui da lam sach o vi tri", (x, y))
        matran[x][y] = '0'
        model[x][y] = '0'
    actions = get_actions(x, y)
    random.shuffle(actions)
    moved = False
    for huong, nx, ny in actions:
        if model[nx][ny] is None:
            print("di chuyen", huong)
            x, y = nx, ny
            moved = True
            break
    if moved == False:
        huong, nx, ny = random.choice(actions)
        print("may hut bui chuyen ngau nhien:", huong)
        x, y = nx, ny
    solan += 1
    print()
    print("buoc", solan, ": di", huong)
    print("ma tran sau khi di chuyen:")
    display()
print("da lam sach sau", solan, "buoc di chuyen.")
```
