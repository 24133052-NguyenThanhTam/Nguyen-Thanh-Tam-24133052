```python
import random
matran = [
    ['1', '0', '1', '1'],
    ['1', '0', '0', '0'],
    ['1', '0', '0', '0'],
    ['1', '0', '1', '1']
]

x, y = 0, 0
def get_action(x,y):
    actions=[]
    if x>0 :actions.append('lên')
    if x<3 :actions.append('xuống')
    if y>0 :actions.append('trái')
    if y<3 :actions.append('phải')
    return actions

def display():
    vitri = [row[:] for row in matran]
    vitri[x][y] = 'R'
    for row in vitri:
        print(' '.join(row))
    print()
print("R là máy hút bụi")
print("1 là bụi")
print("0 là ô đã sạch")
print("Bản đồ ban đầu:")
display()
solan = 0

while '1' in str(matran):
    if matran[x][y] == '1':
        print("Đang hút bụi tại:", x, y)
        matran[x][y] = '0'
    random.shuffle(actions := get_action(x, y))
    for huong in actions:        
        if huong == 'lên' and x > 0:
            x -= 1
        elif huong == 'xuống' and x < 3:
            x += 1
        elif huong == 'trái' and y > 0:
            y -= 1
        elif huong == 'phải' and y < 3:
            y += 1
        solan += 1
        print("Bước", solan, ": đi", huong)
        print("bản đồ sau khi đi chuyển:")
        display()
        break
print("Đã hút sạch bụi sau", solan, "bước.")
```
